 #### Variables in python

```python
i = 28
print(f"i is {i}")

f = 2.8
print(f"f is {f}")

s = "this is a sting"
print(f"s is {s}")

b = True
print(f"b is {b}")

n = None
print(f"n is {n}")
```

- **Interger variable**
  - here i = 28 is an example of integer variable

- **floating point variable**
  - here f = 2.8 is an example of floating point(decimal) variable

- **String variable**
  - here s = "this is a sting" is an example of String variable

- **Boolean variable**
  - here b = True is an example of Boolean variable

- **None Type variable**
  - here n = None, Python also has a particular type called the None type, which only has one possible value, just called a None.


*Note that I never told my program that i is going to be an integer, or f is going to be a floating point number, or b is going to be a Boolean value. Python figures this out for itself. When I say i equals 28, Python knows that if I'm assigning the number 28 to i, then i must be in integer value. And so if you're familiar with other languages like C or Java, where you're used to having to explicitly write out the name of the type for any particular variable, know that Python doesn't require you to do that, and just lets you take a variable and assign it to any given value without needing to specify explicitly what the type of that value actually is.*

## Conditional Statement

```python
x = int(input())

if x > 0:
    print("x is positive")
elif x < 0:
    print("x is negative")
else:
    print("x is 0")
```
